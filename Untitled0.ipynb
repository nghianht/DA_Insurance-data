```python
from scipy import stats
r, p = stats.pearsonr(df.charges, df.age)
r 
```

```python
corr_df = pd.DataFrame(columns=['r','p'])
for col in df:
  if (pd.api.types.is_numeric_dtype(df[col])) & (col != 'charges'):
    r, p = stats.pearsonr(df.charges, df[col])
    corr_df.loc[col] = [round(r, 3), round(p,3)] 
corr_df
```

**Linear Regression**

```python
from scipy import stats
m, p, r, b, err = stats.linregress(df.age, df.charges)

#y = mx + b ~ y =slope(x) + intercept 
print(stats.linregress(df.age, df.charges))

x = range(18,df.age.max())
y = m * x  + b
plt.plot(x, y, color = 'black')

plt.scatter(df.age, df.charges,)
plt.title("Insurance age vs charges")
plt.ylabel("Charges")
plt.xlabel("Age")
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.diagnostic import het_white
from statsmodels.formula.api import ols
#use ols function
model = ols(formula = 'charges~age', data = df).fit()

white_test = het_white(model.resid, model.model.exog)
breusch_test = het_breuschpagan(model.resid, model.model.exog)

output_df = pd.DataFrame(columns = ['LM stats', 'LM p', "F stat", 'F stat p'])
output_df.loc['White'] = white_test
output_df.loc['Breusch'] = breusch_test
output_df
```
